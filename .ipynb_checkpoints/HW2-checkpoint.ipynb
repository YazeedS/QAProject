# 2 Comparing Developers Performance

#### 2.1 Number of bugs per developer

```python
df.Status = df.Status.map({0: 0, 1: "Assigned", 2: "Solved", 3: "Closed"})
```

```python
devs_bugs = pd.DataFrame(columns = ["Dev", "Assigned", "Solved", "Closed", "Reopened"])
```

```python
for i in range(4):
    devs_bugs.loc[i] = [f"Developer {i}",
                        len(df.query(f"Developer == {i} and Status == 'Assigned'")),
                        len(df.query(f"Developer == {i} and Status == 'Solved'")),
                        len(df.query(f"Developer == {i} and Status == 'Closed'")),
                        len(df.query(f"Developer == {i} and Reopn == 1"))]
```

```python
devs_bugs["TotalBugs"] = devs_bugs.Assigned+devs_bugs.Closed + devs_bugs.Solved
```

```python
devs_bugs
```

```python
fig, ax = plt.subplots()


fig.set_size_inches(10, 6);

x_axis = np.arange(4)

# Multi bar Chart

plt.bar(x_axis -0.2, devs_bugs.Assigned, width=0.2, label = 'Assigned', color = "#8B1E3F")
plt.bar(x_axis, devs_bugs.Solved, width=0.2, label = 'Solved', color="#1D7874")
plt.bar(x_axis +0.20, devs_bugs.Closed, width=0.2, label = 'Closed', color = "#B57F50")
plt.bar(x_axis +0.20*2, devs_bugs.Reopened, width=0.2, label = 'Reopened', color="#083D77")

# Ticks
plt.xticks(x_axis+0.1, devs_bugs.Dev);
plt.yticks(np.arange(0, 25, 2));


plt.legend();

fig.tight_layout()

```

```python
np.arange(4)
```

#### 2.2 Average time needed to fix a bug

```python
df.columns
```

```python
df.Dt_Assigned = df.Dt_Assigned.astype("datetime64", "yyyy-mm-dd")
df.Dt_Closed = df.Dt_Closed.astype("datetime64", "yyyy-mm-dd")
df.Dt_Solved = df.Dt_Solved.astype("datetime64", "yyyy-mm-dd")
df.Dt_Report = df.Dt_Report.astype("datetime64", "yyyy-mm-dd")
```

```python
devs_bugs["AvgTime"] = (df.groupby("Developer")["duration"].sum() / df.groupby("Developer")["duration"].count()).values
devs_bugs.AvgTime = devs_bugs.AvgTime.dt.round("h")
```

```python
devs_bugs
```

```python
devs_bugs.AvgTime = devs_bugs.AvgTime.dt.round("h")
```

```python
devs_bugs["floatAvgTime"] = [12.79, 12.33, 12.63, 16.67]
```

```python
fig, ax = plt.subplots()

fig.set_size_inches(10, 6);

x_axis = np.arange(4)

# Multi bar Chart

plt.bar(devs_bugs.Dev, devs_bugs.floatAvgTime, color = ["#8B1E3F", "#1D7874",
                                                                            "#B57F50", "#083D77"])
# Annotations
for i in range(4):
    plt.annotate(f"{devs_bugs.loc[i, 'floatAvgTime']} Days",
                 (i-0.25, devs_bugs.loc[i, "floatAvgTime"]+1), fontsize = 15)

# Ticks

plt.yticks(np.arange(0, 25, 2));

# Titles and labels

plt.title("Average bug closing time")
plt.ylabel("Days")
plt.xlabel("Developers")

fig.tight_layout()

```
